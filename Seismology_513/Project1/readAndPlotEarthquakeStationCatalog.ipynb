```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./networks/iris-gmap/others/iris_allStations.txt', \
                sep='|', \
                comment='#', \
                skipinitialspace=True, \
                dtype= {'Station':str, 'Latitude':float, 'Longitude':float})
```

```python
df.columns = df.columns.str.strip()
```

```python
#plotting the station locations
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(1,1,1)
ax.plot(df['Longitude'], df['Latitude'], '^')
# for i in range(0, df.shape[0]):
#     ax.text(df['Longitude'][i], df['Latitude'][i], df['Station'][i],\
#            color='r')
plt.show()
```

## Earthquake location from NCEDC

```python
import pandas as pd 
import matplotlib as plt

#read the text file with earthquake data
df_eq = pd.read_csv('./events/iris/1973-sep5-23-iris.txt', \
                    sep='|',\
                    skipinitialspace=True, \
                    header=0, skiprows=[1])
```

```python
df_eq.head()
```

```python
df_eq.columns
```

```python
#plotting the data
import matplotlib.pyplot as plt

# Create a figure and a single subplot with a figsize of (6, 6)
fig, ax = plt.subplots(figsize=(6, 6))

# Customize the subplot (ax)
x = df_eq['Lon']
y = df_eq['Lat']
ax.plot(x, y, '.', markersize=1, color='blue', label='Earthquake locations')

# Set labels and title
ax.set_xlabel('Lon')
ax.set_ylabel('Lat')
ax.set_title(f'Earthquake locations map of {df_eq.shape[0]} events')

# Add legend
ax.legend()

# Show the plot
plt.show()
```
